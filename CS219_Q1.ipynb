## Problem 1 c.  

All the routes that pass through the router A2C2 have a loop on them. The traffic is passed back from A2C2 to either A2B1 or A2B2, and comes back to A2C2 and repeats over and over.

One way to ensure that there is no loop, is to make sure that the traffic heading from A2B2/A2B1 to host3 via A2C2, goes to A2C1 instead.
